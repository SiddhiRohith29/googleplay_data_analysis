## Data cleaning

```python
# dropping duplicates from dataframe by taking App column into consideration
proj = proj.drop_duplicates(subset=["App"])
proj
```

```python
# checking for null values
proj.isnull().sum()
```

```python
#filling null values in Rating column with np.mean
def rating_fill(x):
    return x.fillna(x.mean()) 
```

```python
proj.Rating= proj['Rating'].transform(rating_fill)
proj.Rating.isnull().sum()
```

```python
# checking for special characters in Downloads Column
proj['Downloads'].unique()
```

```python
# replacing symbols '+' and ',' and coverting the values of Downloads column into float
proj['Downloads'] = proj.Downloads.str.replace(",","")
proj['Downloads'] = proj.Downloads.str.replace("+","")
proj['Downloads'] = proj['Downloads'].astype(float) # converting the datatype from object to float
```

```python
proj
```

```python
proj.describe()
```

```python
#no. of unique values in Review_count column
proj['Review_count'].unique()
```

```python
# replacing object '0.0' to '0' and '3.0M' to '3000000.0'
proj['Review_count'] = proj.Review_count.replace("0.0",0)
proj['Review_count'] = proj.Review_count.replace("3.0M",3000000.0)
proj['Review_count'] = proj.Review_count.astype(float) # converting datatype to float
```

## Data Visualization

```python
proj.Category.unique()
```

```python
# plotting bar graph and pie chart for column category
sns.catplot(data= proj, kind = 'count', x = 'Category',palette='dark',height=15, aspect=2)
plt.xticks(rotation= 60)
plt.show()
```

```python
#pie chart for detail understanding of Category column
colors = sns.color_palette('pastel')
col_no= proj.Category.value_counts()
plt.pie(col_no, labels =col_no.index, colors = colors, autopct='%.2f%%', radius = 5, shadow= False, startangle=90)
plt.show()
```

```python
# grouping by mean of downloads on each category
cat_down=proj.groupby(['Category'])['Downloads'].mean() 
print(cat_down)
```

```python
#horizontal bar graph for Relation between Category and Downloads
plt.figure(figsize=(12,10))
cat_down.plot(kind="barh")
plt.xlabel('Category')
plt.ylabel('Downloads')
plt.title('Category VS Downloads')
plt.show()
```

```python
# grouping by mean of ratings on each category
cat_rat=proj.groupby(['Category'])['Rating'].mean()
print(cat_rat)
```

```python
#horizontal bar graph for Relation between Category and Rating
plt.figure(figsize=(12,10))
cat_rat.plot(kind="barh")
plt.xlabel('Category')
plt.ylabel('Rating')
plt.title('Category VS Rating')
plt.show()
```

```python
# grouping by mean of ratings on each Download
down_rat=proj.groupby(['Downloads'])['Rating'].mean()
print(down_rat)
```

```python
#horizontal bar graph for Relation between Downloads and Rating
plt.figure(figsize=(12,5))
down_rat.plot(kind="bar")
plt.xlabel('Downloads')
plt.ylabel('Rating')
plt.title('Downloads VS Rating')
plt.show()
```
